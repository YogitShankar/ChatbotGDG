```python
!apt-get update
!apt-get install -y firefox-esr
!pip install selenium
!apt-get install -y xvfb
!pip install pyvirtualdisplay

from pyvirtualdisplay import Display
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import json

# Setup virtual display for Colab
def setup_virtual_display():
    display = Display(visible=0, size=(1024, 768))
    display.start()
    return display

# Function to initialize the Firefox driver with headless mode
def init_firefox_driver():
    options = Options()
    options.headless = True  # Run in headless mode
    driver = webdriver.Firefox(options=options)
    return driver

# Function to store problem statement in a text file
def store_problem_statement(statement, file_name='problem_statement.txt'):
    with open(file_name, 'w', encoding='utf-8') as f:
        f.write(statement)

# Function to store metadata in a JSON file
def store_metadata(metadata, file_name='metadata.json'):
    with open(file_name, 'w', encoding='utf-8') as f:
        json.dump(metadata, f, ensure_ascii=False, indent=4)

# Function to scrape problem data
def scrape_problem(problem_url, driver):
    driver.get(problem_url)

    # Wait for the page to load and check the problem statement
    try:
        print("Waiting for the problem page to load...")
        WebDriverWait(driver, 30).until(
            EC.visibility_of_element_located((By.CLASS_NAME, 'problem-statement'))
        )
        print("Problem page loaded successfully.")
    except Exception as e:
        print(f"Error loading problem page: {e}")
        driver.quit()
        return None

    # Scrape the problem data here (example)
    try:
        title = driver.find_element(By.CLASS_NAME, 'title').text
        statement = driver.find_element(By.CLASS_NAME, 'problem-statement').text
        print(f"Title: {title}")
        print(f"Statement: {statement}")

        # Store the problem statement and metadata
        store_problem_statement(statement)
        store_metadata({'title': title})

        return {'title': title, 'statement': statement}
    except Exception as e:
        print(f"Error scraping problem: {e}")
        driver.quit()
        return None

# Main logic
if __name__ == "__main__":
    # Start virtual display
    display = setup_virtual_display()

    problem_url = "https://codeforces.com/problemset/problem/2053/I1"  # Replace with your desired problem URL
    driver = init_firefox_driver()

    problem_data = scrape_problem(problem_url, driver)

    if problem_data:
        print("Problem Data Scraped Successfully")
        print(problem_data)
    else:
        print("Problem scraping failed.")

    driver.quit()
    # Stop virtual display
    display.stop()
```
